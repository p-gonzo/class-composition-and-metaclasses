```python
class Stack():
    def __init__(self):
        self._storage = {}
        self._count = 0
    
    def push(self, item):
        self._count += 1
        self._storage[str(self._count)]=item
        return self
    
    def pop(self):
        if self._count == 0:
            return 'Stack Empty'
        oldCount = self._count
        self._count -= 1
        return self._storage[str(oldCount)]
    
    def size(self):
        return self._count
    
```

```python
my_stack = Stack()
my_stack.push('a')
my_stack.push('b')
my_stack.push('c')
```
